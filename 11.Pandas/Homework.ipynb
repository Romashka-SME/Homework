## Задание 1
Скачайте с сайта https://grouplens.org/datasets/movielens/ датасет любого размера. Определите какому фильму было выставлено больше всего оценок 5.0.

```python
import pandas as pd
```

```python
data_rate = pd.read_csv('ratings.csv')
data_rate.head()
```

```python
movies = pd.read_csv('movies.csv')
movies.head()
```

```python
merged = data_rate.merge(movies, on='movieId', how='left')
merged.head()
```

```python
merged[merged['rating'] == 5.0]['title'].value_counts().index[0]
```

## Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.

```python
power = pd.read_csv('power.csv')
power.head()
```

```python
power = power[power['quantity'] >= 0]
```

```python
power = power[(power['country'].isin(['Lithuania', 'Latvia', 'Estonia'])) & power['category'].isin([4,12,21])]
```

```python
power = power[power['year'].between(2005,2010)]
```

```python
print(power['quantity'].sum())
```
